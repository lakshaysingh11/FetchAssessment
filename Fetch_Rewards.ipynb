### 2. Second: Write a query that directly answers a predetermined question from a business stakeholder

5. Which brand has the most spend among users who were created within the past 6 months?


```python

# Convert the 'createdDate' column in users_df to datetime
users_df['createdDate'] = pd.to_datetime(users_df['createdDate'])

# Filter users created within the past 6 months
recent_users_df = users_df[users_df['createdDate'] >= pd.Timestamp.now() - pd.DateOffset(months=6)]

# Merge the necessary DataFrames
merged_df = receipts_df.merge(recent_users_df, left_on='userId', right_on='_id')
merged_df = merged_df.merge(brands_df, left_on='rewardsReceiptItemList.brandCode', right_on='brandCode')

# Calculate total spend by brand
brand_spend_df = merged_df.groupby('name')['totalSpent'].sum().reset_index()

# Find the brand with the highest spend
brand_with_highest_spend = brand_spend_df.loc[brand_spend_df['totalSpent'].idxmax()]

# Print the result
brand_name = brand_with_highest_spend['name']
total_spend = brand_with_highest_spend['totalSpent']
print(f"The brand with the highest spend among users created within the past 6 months is {brand_name} with a total spend of ${total_spend}.")
```

```python
#SQL Code
'''
query =
SELECT b.name, SUM(r.totalSpent) AS total_spend
FROM receipts AS r
JOIN users AS u ON r.userId = u._id
JOIN brands AS b ON r.rewardsReceiptItemList.brandCode = b.brandCode
WHERE u.createdDate >= DATE('now', '-6 months')
GROUP BY b.name
ORDER BY total_spend DESC
LIMIT 1
'''
```

6. Which brand has the most transactions among users who were created within the past 6 months?


```python

# Convert the 'createdDate' column in users_df to datetime
users_df['createdDate'] = pd.to_datetime(users_df['createdDate'])

# Filter users created within the past 6 months
recent_users_df = users_df[users_df['createdDate'] >= pd.Timestamp.now() - pd.DateOffset(months=6)]

# Merge the necessary DataFrames
merged_df = receipts_df.merge(recent_users_df, left_on='userId', right_on='_id')
merged_df = merged_df.merge(brands_df, left_on='rewardsReceiptItemList.brandCode', right_on='brandCode')

# Calculate the number of transactions per brand
brand_transactions_df = merged_df['name'].value_counts().reset_index()
brand_transactions_df.columns = ['name', 'transaction_count']

# Find the brand with the most transactions
brand_with_most_transactions = brand_transactions_df.loc[brand_transactions_df['transaction_count'].idxmax()]

# Print the result
brand_name = brand_with_most_transactions['name']
transaction_count = brand_with_most_transactions['transaction_count']
print(f"The brand with the most transactions among users created within the past 6 months is {brand_name} with {transaction_count} transactions.")
```

### 3. Third: Evaluate Data Quality Issues in the Data Provided

We are exploring the "receipts" data provided in the exercise. One common data quality issue to look for is missing or incomplete data. We can check if there are any missing values in important fields such as "totalSpent" and "purchaseDate".

```python
receipts_df = receipts_df.rename(columns={'purchaseDate/$date': 'purchaseDate'})
```

```python
#identify at least one data quality issue.

# Check for missing values in the important fields
missing_total_spent = receipts_df['totalSpent'].isnull().sum()
missing_purchase_date = receipts_df['purchaseDate'].isnull().sum()

# Print the results
print(f"Number of missing values in 'totalSpent': {missing_total_spent}")
print(f"Number of missing values in 'purchaseDate': {missing_purchase_date}")
```
